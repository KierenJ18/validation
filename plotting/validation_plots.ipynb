```python
import awkward as ak

def plot_double_histogram(data1, data2, data3, data4, key, xlabel, ylabel, title_l, title_r, labels, logscale=False):
    """
    Plots two side-by-side histograms for two pairs of datasets.

    Parameters:
        data1, data2: First subplot datasets (e.g., IWCD ID, IWCD OD)
        data3, data4: Second subplot datasets (e.g., HKFD ID, HKFD OD)
        key: The key to extract from each dataset (e.g., "ndigihits")
        xlabel, ylabel, title: Axis labels and main title (title will be used as a prefix)
        labels: List of four labels [label1, label2, label3, label4] for the legend
    """

    
    data1 = ak.to_numpy(ak.flatten(data1[key], axis=None))
    data2 = ak.to_numpy(ak.flatten(data2[key], axis=None))
    data3 = ak.to_numpy(ak.flatten(data3[key], axis=None))
    data4 = ak.to_numpy(ak.flatten(data4[key], axis=None))
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    # First subplot

    axes[0].hist(data1, bins=50, histtype="step", color="blue", linewidth=1.5, label=labels[0])
    axes[0].hist(data2, bins=50, histtype="step", color="red", linewidth=1.5, label=labels[1])

    axes[0].set_xlabel(xlabel)
    axes[0].set_ylabel(ylabel)
    axes[0].set_title(f"{title_l}")
    axes[0].legend()


    axes[1].hist(data3, bins=50, histtype="step", color="blue", linewidth=1.5, label=labels[2])
    axes[1].hist(data4, bins=50, histtype="step", color="red", linewidth=1.5, label=labels[3])
    
    axes[1].set_xlabel(xlabel)
    axes[1].set_title(f"{title_r}")
    axes[1].legend()

    plt.yscale("log" if logscale else "linear")
    plt.tight_layout()
    plt.show()
```

```python
def make_cuts(e_per_trigger, e_per_event, e_cut_params, radial_sq_key="radial_sq", pos_x_key="pos_x", pos_y_key="pos_y", pos_z_key="pos_z", r_cut=325, y_cut=375):
    """
    Make selection cuts for a given particle species.
    
    Returns:
        - pos_x, pos_y, pos_z, pos_r for 0 digit hits (ntriggers==1)
        - pos_x_1, pos_y_1, pos_z_1, pos_r_1 for >=1 digit hits (ntriggers==1)
        - in_tank_r, in_tank_y masks for fiducial volume
    """
    e_digit_hits = e_per_trigger["ndigihits"]
    ntrig_mask = e_per_event["ntriggers"] == 1
    zero_dig_mask = e_digit_hits[ntrig_mask] == 0
    nonzero_dig_mask = np.logical_not(zero_dig_mask)

    pos_x = e_cut_params[pos_x_key][zero_dig_mask]
    pos_y = e_cut_params[pos_y_key][zero_dig_mask]
    pos_z = e_cut_params[pos_z_key][zero_dig_mask]
    pos_r = np.sqrt(e_cut_params[radial_sq_key][zero_dig_mask])

    pos_x_1 = e_cut_params[pos_x_key][nonzero_dig_mask]
    pos_y_1 = e_cut_params[pos_y_key][nonzero_dig_mask]
    pos_z_1 = e_cut_params[pos_z_key][nonzero_dig_mask]
    pos_r_1 = np.sqrt(e_cut_params[radial_sq_key][nonzero_dig_mask])

    in_tank_r = abs(pos_r) <= r_cut
    in_tank_y = abs(pos_y) <= y_cut

    return pos_x, pos_y, pos_z, pos_r, pos_x_1, pos_y_1, pos_z_1, pos_r_1, in_tank_r, in_tank_y
```

```python
def plot_double_2dhistogram(data1, data2, data3, data4, xlabel, ylabel, clabel, title_l, title_r, labels, mask1=False, mask2=False, suptitle=None):

    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    if mask1 is not False:
        data1 = data1[mask1]
        data2 = data2[mask1]
    if mask2 is not False:
        data3 = data3[mask2]
        data4 = data4[mask2]
    # First subplot
    h0 = axes[0].hist2d(data1, data2, bins=(50,50), linewidth=1.5, label=labels[0])
    axes[0].set_xlabel(xlabel)
    axes[0].set_ylabel(ylabel)
    axes[0].set_title(f"{title_l}", size=16)

    # Second subplot
    h1 = axes[1].hist2d(data3, data4, bins=(50,50), label=labels[1])
    axes[1].set_xlabel(xlabel)
    axes[1].set_title(f"{title_r}", size=16)
    plt.grid(False)
    
    # Add colorbars for each subplot
    cbar0 = fig.colorbar(h0[3], ax=axes[0])

    cbar1 = fig.colorbar(h1[3], ax=axes[1])
    cbar1.ax.set_ylabel(clabel)
    
    if suptitle is not None:
        plt.tight_layout(rect=[0, 0, 1, 0.95])
        fig.suptitle(suptitle, fontsize=18, y=0.95)
    else:
        plt.tight_layout()
    plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D  # Import 3D plotting module

# Define cylinder parameters
cylinder_height = 1000
cylinder_radius = 440

# Generate data for the cylinder
z_cylinder = np.linspace(-cylinder_height / 2, cylinder_height / 2, 100)
theta = np.linspace(0, 2 * np.pi, 100)
theta_grid, z_grid = np.meshgrid(theta, z_cylinder)
x_cylinder = cylinder_radius * np.cos(theta_grid)
y_cylinder = cylinder_radius * np.sin(theta_grid)

# Plot the cylinder
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x_cylinder, y_cylinder, z_grid, alpha=0.5, rstride=5, cstride=5, color='blue', edgecolor='k')

# Set axis labels and limits
ax.set_xlabel('X (cm)', fontsize=16)
ax.set_ylabel('Y (cm)', fontsize=16)
ax.set_zlabel('Z (cm)', fontsize=16)
ax.set_xlim(-cylinder_radius, cylinder_radius)
ax.set_ylim(-cylinder_radius, cylinder_radius)
ax.set_zlim(-cylinder_height / 2, cylinder_height / 2)
plt.tight_layout()

plt.show()
```
